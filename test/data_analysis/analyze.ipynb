```python
import matplotlib.pyplot as plt
import numpy as np

test_dirs = ['test_frequent', 'test_somewhat_frequent',
             'test_rare_results', 'test_odd_machines', 'test_machine_6']

fig, ax = plt.subplots(figsize=(10, 6))  # Increase the figure size

for i, test_dir in enumerate(test_dirs):

    scores = []

    with open(f'../test_results/{test_dir}') as f:
        for line in f:
            scores.append(float(line))

    mean = np.mean(scores)
    std = np.std(scores)

    ax.errorbar(i, mean, yerr=std, fmt='.', capsize=5)

    # Add labels for mean and standard deviation with smaller text
    ax.text(i, mean+0.25, f'Mean: {mean:.5f}', ha='center', va='bottom', fontsize=8)
    ax.text(i, mean - 0.5 * std + 0.25, f'Std Dev: {std:.5f}', ha='center', va='top', fontsize=8)

ax.set_title("Test Results Summary")
ax.set_xlabel("Test Case")
ax.set_ylim(bottom=0, top=0.6)
ax.set_ylabel("Time (seconds)")  # Update y label to "Time (seconds)"
ax.set_xticks(range(len(test_dirs)))
ax.set_xticklabels(test_dirs, rotation=45)  # Rotate the labels for better readability
fig.tight_layout()
fig.savefig("test_results_plot.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

test_dirs = ['test_frequent', 'test_somewhat_frequent',
             'test_rare_results', 'test_odd_machines', 'test_machine_6']

fig, ax = plt.subplots()

for i, test_dir in enumerate(test_dirs):

    scores = []

    with open(f'../test_results/{test_dir}') as f:
        for line in f:
            scores.append(float(line))

    mean = np.mean(scores)
    std = np.std(scores)

    ax.errorbar(i, mean, yerr=std, fmt='.', capsize=5)

    # Add labels for mean and standard deviation
    ax.text(i, mean, f'Mean: {mean:.2f}', ha='center', va='bottom')
    ax.text(i, mean - 0.5 * std, f'Std Dev: {std:.2f}', ha='center', va='top')

ax.set_title("Test Results Summary")
ax.set_xlabel("Test Case")
ax.set_ylabel("Mean Score +/- Std Dev")
ax.set_xticks(range(len(test_dirs)))
ax.set_xticklabels(test_dirs, rotation=45)  # Rotate the labels for better readability
fig.tight_layout()
fig.savefig("test_results_plot.png")
plt.show()
```
